```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel("per capita income")
plt.scatter(df.year,df["per capita income (US$)"], color="red", marker = "+")
```

```python
linear_reg = linear_model.LinearRegression()
#why providing year as dataframe
linear_reg.fit(df[["year"]],df["per capita income (US$)"])
```

```python
linear_reg.coef_
```

```python
linear_reg.intercept_
```

```python
linear_reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel("per capita income")
plt.scatter(df.year,df["per capita income (US$)"], color="red", marker = "+")
plt.plot(df["year"],linear_reg.predict(df[["year"]]), color="blue")
```

```python
df["predicted_income"] = linear_reg.predict(df[["year"]])
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel("per capita income")
plt.scatter(df.year,df["per capita income (US$)"], color="red", marker = "+")
plt.plot(df["year"],df["predicted_income"], color="blue")
```
